<h2>5. Data Preparation and Preprocessing</h2>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import PorterStemmer

import warnings
warnings.filterwarnings('ignore')
```

<a id="tokenization"></a>
<h3>5.1. Perform Sentence Tokenization</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
cleaned_sentences = []
```

```python
# Perform sentence tokenization
for i, requirement in enumerate(data.sentence.values):
    token = word_tokenize(requirement)
    cleaned_sentences.append(token)
    if i == 0:
        print('Raw Statement:', end='\n')
        print(requirement, end='\n\n')
        print('Corresponding Tokenized Statement:')
        print(token, end='\n\n')
```

```python
print(cleaned_sentences[:2])
```

<a id="punctuation"></a>
<h3>5.2. Remove Punctuation and Convert all Words to Lowercase</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
for idx, sentence in enumerate(cleaned_sentences):
    temp_sentence = []
    for word in sentence:
        # Convert words in sentences to lowercase
        word = word.lower()
        # Add only alphanumeric words to output
        if word.isalpha():
            temp_sentence.append(word)
    
    # Add lowercase sentences free of punctuations back to cleaned_sentences
    cleaned_sentences[idx] = temp_sentence
```

<a id="stopword"></a>
<h3>5.3. Stopword Removal</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
stop_words = set(stopwords.words('english'))
```

```python
for idx, sentence in enumerate(cleaned_sentences):
    temp_sentence = []
    for word in sentence:
        # Remove all english stopwords
        if word not in stop_words:
            temp_sentence.append(word)
    # Add sentences free of stopwords back to cleaned_sentences
    cleaned_sentences[idx] = temp_sentence
```

<a id="lemmatization"></a>
<h3>5.4 Lemmatization</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
from nltk.stem import WordNetLemmatizer
```

```python
# Create an Instance of the lemmatizer class
lemmatizer = WordNetLemmatizer()
```

```python
for idx, sentence in enumerate(cleaned_sentences):
    temp_sentence = []
    for word in sentence:
        # Perform word lemmatization
        word = lemmatizer.lemmatize(word)
        temp_sentence.append(word)
    # Add sentences with lemmatized words back to cleaned_sentences
    cleaned_sentences[idx] = temp_sentence
```

<a id="word2vec"></a>
<h2>7. Word2Vec Implementation</h2>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
from collections import Counter
from gensim.models import KeyedVectors
```

```python
# Load vectors directly from the file
model = KeyedVectors.load_word2vec_format('Google-News-Vectors/GoogleNews-vectors-negative300.bin', binary=True)
```

<a id="helper_functions"></a>
<h3>7.1. Define Helper Functions</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
def sentence_vector(sentence, word2vec_model):
    # Calculate the vector representation of a sentence
    word_vectors = [word2vec_model[word] for word in sentence.split() if word in word2vec_model]
    if not word_vectors:
        return None
    return np.mean(word_vectors, axis=0)
```

```python
def calculate_similarity(word_list, sentence, word2vec_model):
    # Calculate the similarity between a list of words and a sentence
    word_list_vector = np.mean([word2vec_model[word] for word in word_list if word in word2vec_model], axis=0)
    sentence_vector_ = sentence_vector(sentence, word2vec_model)

    if word_list_vector is None or sentence_vector_ is None:
        return None

    similarity = np.dot(word_list_vector, sentence_vector_) / (np.linalg.norm(word_list_vector) * np.linalg.norm(sentence_vector_))
    return similarity
```

<a id="gather_requirements"></a>
<h3>7.2. Gather Functional Requirements Keywords</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
# Create an empty list for storing requirements keywords
requirement_keywords = list()
```

```python
for requirement in list(data.query('requirement == 1').sentence):
    for text in requirement.split():
        requirement_keywords.append(text)
```

```python
keyword_count = Counter(requirement_keywords)

# Sort the Counter items in descending order by value
sorted_items = sorted(keyword_count.items(), key=lambda x: x[1], reverse=True)

# Take the first 10 values
first_10_values = sorted_items[:10]

# Print the first 10 most frequent functional requirement keywords
print(first_10_values)
```

```python
frequent_keywords = []

for word, frequency in keyword_count.items():
    if frequency >= 3: # Only select words with a frequency greater than or equal to 3
        frequent_keywords.append(word)
```

<a id="gather_requirements"></a>
<h3>7.3. Make Predictions using Test Data</h3>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
preds = []
```

```python
sentences = X_test['sentence']
```

```python
for sentence in sentences:
    similarity = calculate_similarity(requirement_keywords, sentence, model)
    
    if len(sentence.split()) <= 60 and similarity and similarity >= 0.5:
        preds.append(1) # Functional Requirement
    else:
        preds.append(0) # Non Functional Requirements
```

<a id="evaluation_metrics"></a>
<h2>8. Evaluation Metrics</h2>

<a href="#table_of_contents" style="text-decoration: None">Table of Contents</a>

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
print(f"First 10 predictions: {preds[:10]}")

print(f"First 10 actual labels: {list(y_test[:10])}")
```

```python
# Calculate metrics
accuracy = accuracy_score(y_test, preds)
precision = precision_score(y_test, preds)
recall = recall_score(y_test, preds)
f1 = f1_score(y_test, preds)
```

```python
# Create a DataFrame
results_df = pd.DataFrame({
    'Model': ['Word2vec'],
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Recall': [recall],
    'F1-Score': [f1]
})

results_df
```

```python
# Plot confusion matrix
cm = confusion_matrix(y_test, preds)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['True 0', 'True 1'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
